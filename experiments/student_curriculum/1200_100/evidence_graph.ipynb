```python
import matplotlib.pyplot as plt
import sys

f1 = "evidence/10/kl_gibbs.csv" #Vanilla Gibbs
f2 = "evidence/10/nc_kl.csv"
f3 = "evidence/10/block_gibbs.csv"
```

```python
## For variation with time

x1=[]
y1=[]
e1=[]
for l1 in open(f1):
    ln = l1.split(',')
    x1.append(float(ln[0]))
    y1.append(float(ln[1]))
    e1.append(float(ln[3]))
    
x2=[]
y2=[]
e2=[]
for l2 in open(f2):
    ln = l2.split(',')
    x2.append(float(ln[0]))
    y2.append(float(ln[1]))
    e2.append(float(ln[3]))

x3=[]
y3=[]
e3=[]
for l3 in open(f3):
    ln = l3.split(',')
    x3.append(float(ln[0]))
    y3.append(float(ln[1]))
    e3.append(float(ln[3]))
```

```python
lb = max(min(x1),min(x2),min(x3))#,min(x5))
ub = min(max(x1),max(x2),max(x3))#,max(x5))
```

```python
x1_fin = []
y1_fin=[]
e1_fin = []

x2_fin = []
y2_fin=[]
e2_fin = []

x3_fin = []
y3_fin=[]
e3_fin=[]

# x4_fin = []
# y4_fin=[]
# e4_fin=[]

x5_fin = []
y5_fin=[]
e5_fin=[]

#lb=2.5
ub=10
for i in range(len(x1)):
    if x1[i]>=lb and x1[i]<=ub:
        x1_fin.append(x1[i])
        y1_fin.append(y1[i])
        e1_fin.append(e1[i])

for i in range(len(x2)):
    if x2[i]>=lb and x2[i]<=ub:
        x2_fin.append(x2[i])
        y2_fin.append(y2[i])
        e2_fin.append(e2[i])

for i in range(len(x3)):
    if x3[i]>=lb and x3[i]<=ub:
        x3_fin.append(x3[i])
        y3_fin.append(y3[i])
        e3_fin.append(e3[i])

# for i in range(len(x4)):
#     if x4[i]>=lb and x4[i]<=ub:
#         x4_fin.append(x4[i])
#         y4_fin.append(y4[i])
#         e4_fin.append(e4[i])

#for i in range(len(x5)):
#    if x5[i]>=lb and x5[i]<=ub:
#        x5_fin.append(x5[i])
#        y5_fin.append(y5[i])
#        e5_fin.append(e5[i])
#for i in range(int((ub-lb)/10000)+1):
#	x1_fin[i]=x1_fin[i]/1000
#	x2_fin[i]=x2_fin[i]/1000
#	x3_fin[i]=x3_fin[i]/1000
	#x4_fin=x4_fin/1000
#	x5_fin[i]=x5_fin[i]/1000
#plt.yscale('log')    
plt.plot(x1_fin,y1_fin,'b-',marker='d',markersize=9,linewidth=2.0,label = "Vanilla-MCMC")
plt.plot(x2_fin[2:],y2_fin[2:],'g:.',marker='s',markersize=9,linewidth=2.0, label = "VV-MCMC")
plt.plot(x3_fin,y3_fin,'r-.',marker='<',markersize=9,linewidth=2.0, label = "BV-MCMC")
#plt.plot(x4_fin,y4_fin,'m-',marker='>',markersize=18,linewidth=15.0, label = "Vanilla_MOD")
#plt.plot(x5_fin,y5_fin,'k-',marker='*',markersize=18,linewidth=15.0, label = "CON-MCMC("+sys.argv[4]+")")

plt.legend(loc=0,prop={'size':15})

plt.xlabel('Time(s)',fontsize=15, labelpad = 0)
plt.ylabel('KL Div. With True Margs.',fontsize=15, labelpad = 0)

plt.errorbar(x1_fin,y1_fin,yerr=e1_fin,ecolor='b',capsize=5, elinewidth=5)
plt.errorbar(x2_fin[2:],y2_fin[2:],yerr=e2_fin[2:],ecolor='g',capsize=5, elinewidth=5)
plt.errorbar(x3_fin,y3_fin,yerr=e3_fin,ecolor='r',capsize=5, elinewidth=5)
#plt.errorbar(x4_fin,y4_fin,yerr=e4_fin,ecolor='m',capsize=10, elinewidth=5)
#plt.errorbar(x5_fin,y5_fin,yerr=e5_fin,ecolor='k',capsize=10, elinewidth=5)
plt.locator_params(axis='y',nbins=6)
plt.tick_params(axis='x',labelsize=15)
plt.tick_params(axis='y',labelsize=15)
plt.ylim(0,0.04)

plt.title("Job Search - People:50, Evidence:0.1",fontsize=15,y=1.02)
plt.grid(linestyle=':')
#plt.show()
plt.tight_layout()
plt.savefig("job_evidence.pdf", format='pdf')
plt.show()
```
